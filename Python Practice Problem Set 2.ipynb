```python
#problem number 1 
product = str(input("Please Enter your product name : ")) 

#empty sales data 
salesData = []

#enter sales data for each month
for x in range(11):
    unitSales = int(input("Please Enter number of items sold for month " + str(x + 1) + "\t")) 
    while unitSales < 0: 
        unitSales = int(input("Enter a positive value : "))
        
    salesRevenue = int(input("Please Enter total revenue for month " + str(x + 1) + "\t" ))
    while salesRevenue < 0: 
        salesRevenue = int(input("Enter a positive value : ")) 
    
    listItem = [x+1, unitSales, salesRevenue]
    salesData.append(listItem) 
    
print("\n") 
print("Sales Details for " + product)  
print("\n") 
print("--------------------------------------")
print("\n")

monthsForMillion = 1
totalRevenue = 0

for x in range(11):
    month = str((salesData[x])[0])
    unit = str((salesData[x])[1])
    revenue = "{:0,.2f}".format((salesData[x])[2])
    print("Sales Data for month " + month + "\t" + unit + " units" + "\t" + revenue + " USD"  + "\n") 
    
    totalRevenue = totalRevenue + (salesData[x])[2]
    monthsForMillion = monthsForMillion + 1
    
if totalRevenue >= 1000000:
    print("Million Dollars sales revenue has been reached in " + str(monthsForMillion) + " months.")
else:
    revenueSoFar = "{:0,.2f}".format(totalRevenue)
    print("Million Dollars sales revenue has not yet been reached. Your total sales revenue for " + monthsForMillion + " is " + revenueSoFar)
    
```

```python
#problem number 2 
loanAmount = float(input("Enter the total loan amount : "))
while loanAmount < 0: 
        loanAmount = int(input("Enter a positive value : ")) 
        
loanPeriod = float(input("Enter the loan period in months : "))
while loanPeriod < 0: 
        loanPeriod = int(input("Enter a positive value : ")) 
        
annualIR = float(input("Enter the annual interest rate : ")) 
while annualIR < 0: 
        annualIR = int(input("Enter a positive value : ")) 

monthlyPayment = 0.00
totalPayment = 0.00
totalInterest = 0.00 

#calculations 
monthlyIR = annualIR/1200
a = loanAmount * monthlyIR
c = pow((1 + monthlyIR), loanPeriod)
d = (1 - (1/c))

monthlyPayment = a/d
totalPayment = monthlyPayment * loanPeriod
totalInterest = totalPayment - loanAmount 

print("\n")

#show outputs
monthlyPaymentStr = "{:0,.2f}".format(monthlyPayment)
print("Monthly payment is : \t" + monthlyPaymentStr)
print("\n")

totalPaymentStr = "{:0,.2f}".format(totalPayment)
print("Total Payment is : \t" + totalPaymentStr)
print("\n")

totalInterestStr = "{:0,.2f}".format(totalInterest)
print("Total Interest is : \t" + totalInterestStr)
print("\n")

```

```python
#problem number 3 
rangers = []
yankees = []

print("Enter points for Rangers:")
for i in range(7):
    print("Game", i+1, "Points")
    innings = []

    for j in range(9):
        innings.append(int(input("Enter score for innings"+ str(j+1)+": " )))
        rangers.append(innings)
    

print("Enter points for Yankees:")
for i in range(7):
    print("Game", i+1, "Points")
    innings = []
  
    for j in range(9):
        innings.append(int(input("Enter score for innings"+ str(j+1)+": " )))
        yankees.append(innings)

for i in range(7):
    r = 0
    y = 0

    for j in range(9):
        r = r+rangers[i][j]
        y = y+yankees[i][j]

    if(r>y):
        print("Game "+str(i+1)+": Rangers wins.")
    elif(y>r):
        print("Game "+str(i+1)+": Yankees wins.")
    else:
        print("Game "+str(i+1)+": Both the teams tied at the end of the 9th inning.")
```
